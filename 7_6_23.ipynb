```python
#mysql connection
import mysql.connector
connection=mysql.connector.connect(
   host="localhost",
   user="root",
   password="SATHYAs123$>>",
   database="c361cohort"
)
cursor=connection.cursor()

create_table_query="""CREATE TABLE t1(
id int primary key auto_increment,
c1 varchar(20),
c2 varchar(20)
)"""


cursor.execute(create_table_query)
print("Table created successfully")

insert_query = "INSERT INTO t1 (c1, c2) VALUES (%s, %s)"

#tuples in list
data = [
  ('A', 'v1'),
  ('B', 'v2'),
  ('C', 'v3'),
  ('D', 'v4'),
  ('E', 'v5'),
  ('F', 'v6')
]

#executemany() is tuples in list 
#execute() if tuple 
cursor.executemany(insert_query,data)

connection.commit()

select_query="select * from t1"

cursor.execute(select_query)

rows=cursor.fetchall()

for row in rows:
   print(row)
   
cursor.close()
connection.close()

```
